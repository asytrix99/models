```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function(func, x_range=(-10, 10), num_points=1000, title=None, 
                  xlabel='x', ylabel='y', grid=True, figsize=(10, 6)):
    """
    Plot a mathematical function.
    
    Parameters:
    -----------
    func : callable
        A function that takes a numpy array and returns a numpy array
    x_range : tuple, optional
        The range of x values as (x_min, x_max). Default is (-10, 10)
    num_points : int, optional
        Number of points to plot. Default is 1000
    title : str, optional
        Title of the plot. If None, uses the function name
    xlabel : str, optional
        Label for x-axis. Default is 'x'
    ylabel : str, optional
        Label for y-axis. Default is 'y'
    grid : bool, optional
        Whether to show grid. Default is True
    figsize : tuple, optional
        Figure size as (width, height). Default is (10, 6)
    
    Returns:
    --------
    fig, ax : matplotlib figure and axis objects
    
    Example:
    --------
    >>> import numpy as np
    >>> plot_function(np.sin, x_range=(-2*np.pi, 2*np.pi), title='Sine Wave')
    >>> plt.show()
    
    >>> # Custom function
    >>> def my_func(x):
    ...     return x**2 - 4*x + 3
    >>> plot_function(my_func, x_range=(-2, 6), title='Quadratic Function')
    >>> plt.show()
    """
    # Generate x values
    x = np.linspace(x_range[0], x_range[1], num_points)
    
    # Calculate y values
    try:
        y = func(x)
    except Exception as e:
        raise ValueError(f"Error evaluating function: {e}")
    
    # Create the plot
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot(x, y, linewidth=2, color='blue')
    
    # Set title
    if title is None:
        title = f'Graph of {func.__name__}(x)'
    ax.set_title(title, fontsize=14, fontweight='bold')
    
    # Set labels
    ax.set_xlabel(xlabel, fontsize=12)
    ax.set_ylabel(ylabel, fontsize=12)
    
    # Add grid
    if grid:
        ax.grid(True, alpha=0.3)
    
    # Add axes lines at y=0 and x=0
    ax.axhline(y=0, color='k', linewidth=0.5)
    ax.axvline(x=0, color='k', linewidth=0.5)
    
    return fig, ax
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Plot your own function
def my_func(x):
    return np.abs(2*x+2)

plot_function(my_func, x_range=(-10, 10))
plt.show()
```
