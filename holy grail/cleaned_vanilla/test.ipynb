```python
from sympy import symbols, And, Not, Or, Implies, Equivalent
from sympy.logic.boolalg import truth_table
from sympy.logic.inference import satisfiable

A,B,C,D = symbols('A,B,C,D')

Sa = Implies(A,~B)
Sb = Implies(~A,B)
Sc = Implies(Or(~B,D),D)
Sd = Implies(C,A)

KB = And(Sa, Sb, Sc, Sd)

# for assignment, truth_value in truth_table(KB, [A, B, C]):
#     print(f"[{A} = {'True' if assignment[0] else 'False'}, {B} = {'True' if assignment[1] else 'False'}, {C} = {'True' if assignment[2] else 'False'}] KB={'True' if truth_value else 'False'}")

def contradiction_check(KB, query):
    return not satisfiable(And(KB, Not(query)))
```
