```python
from scipy import stats

mu = 30 # mean
sigma = 1.8 # s.d.
x1 = 28 # lower bound
x2 = 33 # upper bound

# calcualte probabilities
p_lower = stats.norm.cdf(x1,mu,sigma) # prob from z=0 to upper bound
p_upper = stats.norm.cdf(x2,mu,sigma) # prob from z=0 to lower bound

# probability of interval
prob = p_upper - p_lower

# print results
print('\n')
print(f'normal distribution: mean = {mu}, std dev = {sigma} \n')
print(f'probability of occuring btw {x1} & {x2}: ')
print(f'--> inside interval pin = {round(prob*100,1)}%')
print(f'--> outside interval Pout = {round((1-prob)*100,1)}%\n')
print('\n')
```
