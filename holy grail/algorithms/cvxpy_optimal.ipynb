```python
import cvxpy as cp

x1 = cp.Variable()
x2 = cp.Variable()
x3 = cp.Variable()
x4 = cp.Variable()
x5 = cp.Variable()

objective = cp.Maximize(2*x1-3*x2+x3)

constraints = [
    x1-x2+x3<=5,
    x1-x2+4*x3<=7,
    x1+2*x2-x3+x4<=14,
    x3-x4+x5<=7,
    x1>=-15,x2>=-15,x3>=-15,x4>=-15,x5>=-15,
    x1<=15,x2<=15,x3<=15,x4<=15,x5<=15
]

problem = cp.Problem(objective, constraints)
problem.solve()

print(f"Status: {problem.status}")
print(f"Optimal value: {problem.value}")
print(f"Optimal a: {x1.value}")
print(f"Optimal b: {x2.value}")
print(f"Optimal b: {x3.value}")
print(f"Optimal b: {x4.value}")
print(f"Optimal b: {x5.value}")
```
