```python
# Define a function to perform Breadth-First Search (BFS)
# and find the shortest path from 'start' to 'goal'.
def bfs_shortest_path(graph, start, goal):
    visited = []  # List to keep track of cities we have already visited
    queue = []    # Queue for BFS: stores cities to explore next
    path = {}  # Dictionary to remember how we got to each city

    queue.append(start)   
    visited.append(start) 
    path[start] = None    # start node has no predecessor

    # Loop until there are no more cities to explore
    while queue:
        city = queue.pop(0) # Get the next city to explore (FIFO order)

        # If we've reached the goal, reconstruct the path and return it
        if city == goal:
            result = []  # List to store the path from start to goal
            while city is not None:
                result.append(city) 
                city = path[city] 
            result.reverse()  # reverse to get path from start to goal
            return result 
        
        # Check all neighboring cities connected to the current city
        for neighbor in graph[city]:
            if neighbor not in visited:
                visited.append(neighbor) 
                queue.append(neighbor)    
                path[neighbor] = city   # Record that we came to 'neighbor' from 'city'

    return None  # if no path found

# Define the graph using a dictionary.
# Each city (node) maps to a list of cities it has direct flights to (neighbors).

# CHANGE THIS
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D', 'E'],
    'C': ['A', 'B', 'F', 'I'],
    'D': ['B', 'G'],
    'E': ['B', 'G', 'H'],
    'F': ['C', 'H'],
    'G': ['D', 'E', 'S'],
    'H': ['E', 'F', 'J', 'S'],
    'I': ['C', 'J'],
    'J': ['I', 'K'],
    'K': ['J', 'S'],
    'S': ['G', 'H']
}
# Find the shortest path using bfs
route = bfs_shortest_path(graph, 'A', 'S')
# CHANGE THIS

if route:
    print("Shortest path:", route)
    print("Min # hops:", len(route) - 1)
else:
    print("No route found.")
```
