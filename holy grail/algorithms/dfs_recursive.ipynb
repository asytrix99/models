```python
# Recursive DFS function
def dfs(graph, node, visited):
    print(node)
    visited.append(node)

    for neighbour in graph[node]:
        if neighbour not in visited:
            dfs(graph, neighbour, visited)

# CHANGE THIS
graph = {
    'a': ['b','c'],
    'b': ['a','d','e'],
    'c': ['a','f'],
    'd': ['b'],
    'e': ['b'],
    'f': ['c']
}
visited = []
dfs(graph, 'a', visited)
# CHANGE THIS

print("\nVisit_sequence:", visited)
```
